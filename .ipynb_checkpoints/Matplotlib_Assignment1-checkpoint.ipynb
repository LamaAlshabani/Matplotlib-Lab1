```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
data = {'name': ['Dana', 'Layan', 'Alaa', 'Lama'],
        'result': [60, 80, 90, 100]}
DF = pd.DataFrame(data, columns=["name","result"])


plt.plot(DF.name,DF.result, c ='green');
plt.plot(['Dana', 'Layan', 'Alaa', 'Lama'], [20,30,40,50])


plt.xlabel('name')
plt.ylabel('result')
plt.title('The result of 4 students in two exams');



```

## Q2: Create pie chart with equal explode values.

```python
name = DF.name
pr =DF.result
plt.pie(pr,labels=name, autopct='%1.2f%%', radius=1, explode = [0.1,0.1,0.1,0.1]);
plt.title("Percentage of Students' results in the first exam");
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
third=plt.scatter(DF.name,DF.result, alpha = 0.4 , marker = 'o');
plt.xlabel('name')
plt.ylabel('result')
plt.title('The result of 4 students in the first exam');
plt.figure(figsize=(50,50));
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(DF.name,DF.result, alpha = 0.4 , marker = 'o');
plt.xlabel('name',fontsize = 16)
plt.ylabel('result',fontsize = 16)
plt.title("The result of 4 students in the first exam"  , c="r",loc = 'left');
plt.figure(figsize=(50,50));
```

## Q5: Create three line charts with Legends at left and colors.

```python
plt.plot(DF.name,DF.result, c ='green');
plt.plot(['Dana', 'Layan', 'Alaa', 'Lama'], [20,30,40,50], c="r")
plt.plot(['Dana', 'Layan', 'Alaa', 'Lama'], [50,70,80,90], c="b")
plt.legend(["first exam","second exam","third exam"],loc = 'upper left')
plt.xlabel('name')
plt.ylabel('result')
plt.title('The result of 4 students in three exams');
```
